```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for points per game averages
points_data = {
    'Rookie of the Year': ['LaMelo Ball', 'Ja Morant', 'Luka Doncic', 'Ben Simmons', 'Malcolm Brogdon', 'Karl-Anthony Towns', 'Andrew Wiggins', 'Michael Carter-Williams', 'Damian Lillard', 'Kyrie Irving'],
    'Points per game': [15.7, 17.8, 21.2, 15.8, 10.2, 18.3, 16.9, 16.7, 19.0, 18.5]
}
points_df = pd.DataFrame(points_data)

# Data for assists per game averages
assists_data = {
    'Rookie of the Year': ['LaMelo Ball', 'Ja Morant', 'Luka Doncic', 'Ben Simmons', 'Malcolm Brogdon', 'Karl-Anthony Towns', 'Andrew Wiggins', 'Michael Carter-Williams', 'Damian Lillard', 'Kyrie Irving'],
    'Assists per game': [6.1, 7.3, 6.0, 8.2, 4.2, 2.3, 2.1, 6.3, 6.5, 5.4]
}
assists_df = pd.DataFrame(assists_data)

# Data for rebounds per game averages
rebounds_data = {
    'Rookie of the Year': ['LaMelo Ball', 'Ja Morant', 'Luka Doncic', 'Ben Simmons', 'Malcolm Brogdon', 'Karl-Anthony Towns', 'Andrew Wiggins', 'Michael Carter-Williams', 'Damian Lillard', 'Kyrie Irving'],
    'Rebounds per game': [5.9, 3.9, 7.8, 8.1, 4.2, 10.5, 4.6, 6.2, 3.1, 3.7]
}
rebounds_df = pd.DataFrame(rebounds_data)

# Data for steals per game averages
steals_data = {
    'Rookie of the Year': ['LaMelo Ball', 'Ja Morant', 'Luka Doncic', 'Ben Simmons', 'Malcolm Brogdon', 'Karl-Anthony Towns', 'Andrew Wiggins', 'Michael Carter-Williams', 'Damian Lillard', 'Kyrie Irving'],
    'Steals per game': [1.6, 0.9, 1.0, 1.7, 0.9, 0.7, 1.0, 1.9, 0.9, 1.1]
}
steals_df = pd.DataFrame(steals_data)

# Data for blocks per game averages
blocks_data = {
    'Rookie of the Year': ['LaMelo Ball', 'Ja Morant', 'Luka Doncic', 'Ben Simmons', 'Malcolm Brogdon', 'Karl-Anthony Towns', 'Andrew Wiggins', 'Michael Carter-Williams', 'Damian Lillard', 'Kyrie Irving'],
    'Blocks per game': [1.6, 0.3, 0.3, 0.9, 0.2, 1.7, 0.6, 0.6, 0.4, 0.4]
}
blocks_df = pd.DataFrame(blocks_data)
```

```python
# Creates a bar chart for points per game averages
plt.bar(points_df['Rookie of the Year'], points_df['Points per game'])
plt.title('Points per game averages of Rookie of the Year players (last 10 years)')
plt.xlabel('Rookie of the Year')
plt.ylabel('Points per game')
plt.xticks(rotation=90)
plt.show()

# Creates a bar chart for assists per game averages
plt.bar(assists_df['Rookie of the Year'], assists_df['Assists per game'])
plt.title('Assists per game averages of Rookie of the Year players (last 10 years)')
plt.xlabel('Rookie of the Year')
plt.ylabel('Assists per game')
plt.xticks(rotation=90)
plt.show()

# Creates a bar chart for rebounds per game averages
plt.bar(rebounds_df['Rookie of the Year'], rebounds_df['Rebounds per game'])
plt.title('Rebounds per game averages of Rookie of the Year players (last 10 years)')
plt.xlabel('Rookie of the Year')
plt.ylabel('Rebounds per game')
plt.xticks(rotation=90)
plt.show()

# Creates a bar chart for steals per game averages
plt.bar(steals_df['Rookie of the Year'], steals_df['Steals per game'])
plt.title('Steals per game averages of Rookie of the Year players (last 10 years)')
plt.xlabel('Rookie of the Year')
plt.ylabel('Steals per game')
plt.xticks(rotation=90)
plt.show()

# Creates a bar chart for rebounds per game averages
plt.bar(blocks_df['Rookie of the Year'], blocks_df['Blocks per game'])
plt.title('Blocks per game averages of Rookie of the Year players (last 10 years)')
plt.xlabel('Rookie of the Year')
plt.ylabel('Blocks per game')
plt.xticks(rotation=90)
plt.show()
```
